```python
from google.colab import files
import fitz  # PyMuPDF
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Upload PDFs
uploaded_files = files.upload()
```

```python
def extract_text_from_pdf(file_bytes):
    doc = fitz.open(stream=file_bytes, filetype="pdf")
    text = ""
    for page in doc:
        text += page.get_text("text")  # try explicit text extraction mode
    return text.strip()


# Extract and show the text content of each PDF
pdf_texts = []
resume_names = []

print("\n📄 Extracted Resume Texts:\n")
for filename, file in uploaded_files.items():
    text = extract_text_from_pdf(file)
    pdf_texts.append(text)
    resume_names.append(filename)
    print(f"--- {filename} ---\n")
    print(text[:1000])  # Print first 1000 chars for brevity
    print("\n" + "="*50 + "\n")
```

```python
# Load SentenceTransformer model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Define job description


job_description = """ We are looking for a Python Developer with experience in Flask, REST APIs,
and working knowledge of databases like PostgreSQL. Familiarity with Docker and Git is a plus.
"""

```

```python
# Function to compute similarity
def compute_similarity(job_description: str, resumes: list, names: list) -> list:
    job_embedding = model.encode([job_description])
    resume_embeddings = model.encode(resumes)
    scores = cosine_similarity(job_embedding, resume_embeddings)[0]
    scored_resumes = list(zip(names, resumes, scores))
    scored_resumes.sort(key=lambda x: x[2], reverse=True)
    return scored_resumes

# Compute similarity
results = compute_similarity(job_description, pdf_texts, resume_names)

# Show ranked results
print("\n📊 Resume Ranking by Similarity:\n")
for i, (name, resume_text, score) in enumerate(results, 1):
    print(f"{i}. {name} — Similarity Score: {score:.2f}")
```
